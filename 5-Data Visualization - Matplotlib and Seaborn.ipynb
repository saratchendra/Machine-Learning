### Histogram

#### A histogram is a graphical display of data using bars of different heights. In a histogram, each bar groups numbers into ranges. Taller bars show that more data falls in that range.

```python
x=[1,2,3,4,6]
plt.hist(x)
```

```python
x= np.array(range(10,100, 5))
plt.hist(x)
plt.title('Histogram')
```

```python
x= np.array(range(10,100, 5))
plt.hist(x, bins=25)           # bins - to create intervel
plt.title('Histogram')
```

### Boxplot

#### The box and whisker plot are the best way to find the central tendency(median, mean, and mode). Used for anomaly detection(outliers). Used to find the outliers. Used for inter quartile range.
Maximum: Q3 + 1.5*IQR

Minimum: Q1 -1.5*IQR

```python
data=[70,150,160,215,300,500]
plt.boxplot(data );
```

```python
# From the plot it is observed that 70 is min, 300 is max, 160 is median, 500(circle) is the outlier.
```

```python
data=[25, 29, 30, 35, 40,50,55,60, 100]
plt.boxplot(data );
```

```python
# From the above graph it is observed that min=25, max=60, median=40, outlier=100(in circle)
# outlier - The value which is beyond the range either low or high
```

```python
data=[1, 30,40,45,50,55,60,65]
plt.boxplot(data);
```

```python
# from the above graph outlier=1
```

### Pie chart

#### A pie chart is a circular statistical graphic, which is divided into slices to illustrate numerical proportion. In a pie chart, the arc length of each slice, is proportional to the quantity it represents.

```python
food='Dominos', 'KFC', 'Burger king', 'Subway'
income_dollars = [2500, 3000, 1500, 1300 ]
explode = (0,0.3,0,0) # explode second slice i.e., KFC
plt.pie(income_dollars, explode=explode, labels=food,autopct='%1.1f%%', shadow=False);
```

## Seaborn
Seaborn is a Python data visualization library based on matplotlib. It provides a high-level interface for drawing attractive and informative statistical graphics

```python
# Load library
import seaborn as sns
```

#### Types of plots
1) relplot(),
2) lineplot(),
3) scatterplot(),
4) boxplot(),
5) catplot(),
6) regplot(),
7) lmplot(),
8) jointplot(),
9) distplot(),
10) violinplot
11) pairplot


```python
# The plots have been divided for categorical and numerical variables
```

```python
# countplot()  - used for nominal and small ordinal categorical data.
# kdeplot()    - used for interval data.
# jointplot()  - used for interval and some nominal categorical data.
```

```python
# seaborn contains many in-built datasets which can be easily loaded
df=sns.load_dataset('tips')
```

```python
df.head()   # head - to get first 5 rows information
```

```python
# countplot() - for categorical variable
sns.countplot(x='sex', data=df)
```

```python
sns.countplot(x='smoker', data=df)
```

```python
sns.countplot(x='time', data=df)
```

```python
sns.countplot('size', data=df)
```

```python
# Always make an analysis from the graphs to know more information. 
```

#### distplot() - Seaborn distplot lets you show a histogram with a line on it. This can be shown in all kinds of variations. We use seaborn in combination with matplotlib, the Python plotting module.

```python
sns.distplot(df['total_bill'])
```

```python
sns.distplot(df['tip'])
```

```python
# bins(x) - bins/intervels can also be added
```

```python
sns.distplot(df['tip'],kde=False,bins=10)
```

```python
# scatterplot() - for numerical variable
sns.scatterplot(x='total_bill', y='tip',data=df)
```

```python
# from the abpve graph it is observed that tip incresed with the total_bill. 
# there is a correlation between total_bill and tip.
```

### Heatmap - it is used to find the correlation between the variables

```python
heatmat=sns.heatmap(df.corr(), cmap='BuGn', linewidth=.5, annot=True)
heatmat
```

```python
df.corr()
```

```python
# total_bill to total_bill has 1 correlation
# total_bill to tip has 0.67 correlation i.e., tip=0.67*total_bill
# total_bill to size has 0.59 correlation i.e., size=0.59*total_bill
# it follows same with tip to total_bill and size. 
```

```python
sns.boxplot(x="total_bill", y="day", hue="smoker",palette='rainbow', data=df)
```

```python
# from the above figure it is observed that people visit more on sundays than any other day smoke alot.
```

```python
sns.violinplot(x="total_bill", y="day", data=df)
```
